```python
student_marks = {'Priya': 85, 'Deepak': 72, 'Siddharth': 78}
for student,marks in student_marks.items():
  if student == 'Siddharth':
    if  marks >80:
      print(student, ':' ,'Distinction')
    elif marks >60 and marks<=80:
      print(student, ':' ,'First Class')
    else:
      print(student, ':' ,'Pass')

```

```python
student_marks = {'Priya': 85, 'Deepak': 72, 'Siddharth': 78}
stu_name = 'Siddharth'
if stu_name in student_marks:
  marks = student_marks[stu_name]
  if marks >80:
    print(f"{stu_name} : Distinction")
  elif marks >60 and marks<=80:
    print(f"{stu_name} : First Class")
  else:
    print(f"{stu_name} : Pass")
```
