### Q1: The Security Gate (While Loop)
Create a login system that:
- Asks the user for a password input.
- Gives the user exactly **3 attempts**.
- If the password is `Inceptez2026`, print "Access Granted" and exit the loop.
- If the password is wrong, print "Incorrect password. X attempts remaining."
- After 3 failed attempts, print "Account Locked: Please contact admin."

```python
# Write your logic here
password_correct = "Inceptez2026"
attempts = 3 # Loop Initialization
while attempts >0: # Loop Condition
  password = input("Enter Password: ")
  if password == password_correct:
    print("Access Granted")
    break
  else:
    attempts -= 1 # Loop Increment or Decrement.It is bitwise operator.
    print(f"Incorrect password. {attempts} remaining.")
  if attempts == 0:
    print("Account Locked: Please contact admin.")
```

### Q2: Filtered Iteration (Break & Continue)
Write a program that iterates through a range of numbers from 1 to 30:
- If the number is divisible by 5, **skip** it using `continue`.
- If the number is 27, **stop** the loop entirely using `break`.
- Print all other numbers on the same line.

```python
# Write your code here
for num in range(1,31):
  if num == 27:
    break
  elif num % 5 == 0:
    continue
  else:
    print(num,end = ' ')
```

```python
numbers = []
for num in range(1,31):
  if num == 27:
    break
  elif num % 5 == 0:
    continue
  else:
    numbers.append(str(num))
print(' '.join(numbers)) # joins iterate a string on other word we need data structure
```
